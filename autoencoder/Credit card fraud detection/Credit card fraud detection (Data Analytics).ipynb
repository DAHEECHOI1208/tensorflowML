```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

df = pd.read_csv('./creditcard.csv')
df.head()
```

```python
df.describe()
```

## Transaction per Time between fraud and non-fraud

```python
#create figure that has 2 graph (in a column), it shares X axis
f,(ax1,ax2) = plt.subplots(2,1,sharex=True,figsize=(12,4))
bins = 100 # 가로의 버킷수.

# draw frauded graph. X is time, Y is amount of transaction for the time
ax1.hist(df.Time[df.Class==1],bins = bins)
ax1.set_title('Fraud')

ax2.hist(df.Time[df.Class==0],bins = bins)
ax2.set_title('Non-Fraud')

plt.xlabel('Time (in Seconds)')
plt.ylabel('Number of Transactions')
plt.show()
```

## Transaction amount comparison

```python
f,(ax1,ax2) = plt.subplots(2,1,sharex=True,figsize=(12,4))
bins = 40

ax1.hist(df.Amount[df.Class==1],bins=bins)
ax1.set_title('Fraud')

ax2.hist(df.Amount[df.Class==0],bins=bins)
ax2.set_title('Non-Fraud')

plt.xlabel('Amount($)')
plt.ylabel('Number of transaction')
plt.yscale('log')
plt.show()
```

```python
features = df.ix[:,1:29].columns;

plt.figure(figsize=(12,28*4))
gs = gridspec.GridSpec(28,1)
for i,cn in enumerate(df[features]):
   
    ax = plt.subplot(gs[i])
    sns.distplot(df[cn][df.Class == 1], bins=50,color='r')
    sns.distplot(df[cn][df.Class == 0], bins=50,color='b')
    ax.set_xlabel('')
    ax.set_title('histogram of feature: ' + str(cn))

plt.show()
```
